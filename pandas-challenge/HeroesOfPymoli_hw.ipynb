```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
pymoli_data = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
pymoli_purchases_df = pd.read_csv(pymoli_data)

pymoli_purchases_df
```

## Player Count

* Display the total number of players


```python
total_players_df = len(pymoli_purchases_df["SN"].unique())

total_number_of_players_df = pd.DataFrame({"Total Players": [total_players_df]})

total_number_of_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(pymoli_purchases_df["Item ID"].unique())


average_price = pymoli_purchases_df["Price"].mean()


total_purchases = len(pymoli_purchases_df["Purchase ID"].unique())


total_revenue = pymoli_purchases_df["Price"].sum()


purchase_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                              "Average Price": [average_price],
                              "Number of Purchases": [total_purchases],
                              "Total Revenue": [total_revenue]})

purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map("${:.2f}".format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("${:.2f}".format)

purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed


```python
pymoli_purchases_unique_df = pymoli_purchases_df.drop_duplicates("SN")

pymoli_purchases_unique_df

gender_demos_df = pymoli_purchases_unique_df["Gender"].value_counts()

gender_demos_df

gender_total_df = pymoli_purchases_unique_df["Gender"].count()

gender_total_df

gender_percent_df = gender_demos_df / gender_total_df * 100

gender_percent_df

gender_demographics_df = pd.DataFrame({"Total Count": gender_demos_df,
                              "Percent of Players": gender_percent_df})

gender_demographics_df["Percent of Players"] = gender_demographics_df["Percent of Players"].map("{:.2f}%".format)

gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_groups_df = pymoli_purchases_df.groupby("Gender")

gender_average_purchase_df = gender_groups_df["Price"].mean()

gender_average_purchase_df

gender_total_purchase_df = gender_groups_df["Price"].sum()

gender_total_purchase_df

gender_total_by_person_df = gender_total_purchase_df / gender_demos_df

gender_total_by_person_df


gender_analysis_df = pd.DataFrame({"Purchase Count": pymoli_purchases_df["Gender"].value_counts(),
                              "Average Purchase Price": gender_average_purchase_df,
                            "Total Purchase Value": gender_total_purchase_df,
                                      "Avg. Total Purchase per Person": gender_total_by_person_df})

gender_analysis_df["Average Purchase Price"] = gender_analysis_df["Average Purchase Price"].map("${:.2f}".format)
gender_analysis_df["Total Purchase Value"] = gender_analysis_df["Total Purchase Value"].map("${:.2f}".format)
gender_analysis_df["Avg. Total Purchase per Person"] = gender_analysis_df["Avg. Total Purchase per Person"].map("${:.2f}".format)

gender_analysis_df
```

```python
# Data Observation 1 (Gender)

# Although there is a disparaging difference in the gender purchase count between male and female, 
# the lesser amount of female purchases allowed for greater price value than the males. 
# This is proven by the 18 cent difference in average purchase price, 
# in addition to the average total purchase per person being 40 cents higher. Males are the clearly
# the primary gender demographic with interest in the games, but the females produced higher average purchases,
# thus showing that their market is formidable despite having less players in comparison to males.
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
pymoli_purchases_unique_df = pymoli_purchases_df.drop_duplicates("SN")

pymoli_purchases_unique_df

age_groups_df = pymoli_purchases_unique_df.groupby("Age")

bins = [6.99, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 45]

group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


pymoli_purchases_unique_df['Age Group'] = pd.cut(pymoli_purchases_unique_df['Age'], bins, labels=group_labels)
pymoli_purchases_unique_df.head()


age_demos_df = pymoli_purchases_unique_df["Age Group"].value_counts()

age_demos_df

player_total_df = pymoli_purchases_unique_df["Age"].count()

player_total_df

age_percent_df = age_demos_df / len(pymoli_purchases_unique_df) * 100

age_percent_df

age_demographics_df = pd.DataFrame({"Total Count": age_demos_df,
                                    "Percent of Players": age_percent_df})

age_demographics_df["Percent of Players"] = age_demographics_df["Percent of Players"].map("{:.2f}%".format)

age_demographics_df = age_demographics_df.sort_index()

age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pymoli_purchases_unique_df = pymoli_purchases_df.drop_duplicates("SN")

pymoli_purchases_unique_df

age_groups_df = pymoli_purchases_df.groupby("Age")

bins = [6.99, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 45]

group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

pymoli_purchases_df['Age Group'] = pd.cut(pymoli_purchases_df['Age'], bins, labels=group_labels)
pymoli_purchases_df


age_group_df = pymoli_purchases_df.groupby("Age Group")

age_average_purchase_df = age_group_df["Price"].mean()

age_average_purchase_df

age_total_purchase_df = age_group_df["Price"].sum()

age_total_purchase_df

age_total_by_person_df = age_group_df['Price'].count()

age_total_by_person_df

average_total_by_person_df = age_total_purchase_df / age_demographics_df["Total Count"]

average_total_by_person_df


age_analysis_df = pd.DataFrame({"Purchase Count": pymoli_purchases_df["Age Group"].value_counts(),
                                "Average Purchase Price": age_average_purchase_df,
                                "Total Purchase Value": age_total_purchase_df,
                                "Avg. Total Purchase per Person": average_total_by_person_df})

age_analysis_df["Average Purchase Price"] = age_analysis_df["Average Purchase Price"].map("${:.2f}".format)
age_analysis_df["Total Purchase Value"] = age_analysis_df["Total Purchase Value"].map("${:.2f}".format)
age_analysis_df["Avg. Total Purchase per Person"] = age_analysis_df["Avg. Total Purchase per Person"].map("${:.2f}".format)

age_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
return_players_df = pymoli_purchases_df.groupby("SN")

return_average_purchase_df = return_players_df["Price"].mean()

return_average_purchase_df

return_total_purchase_df = return_players_df["Price"].sum()

return_total_purchase_df

top_spenders_df = pd.DataFrame({"Purchase Count": pymoli_purchases_df["SN"].value_counts(),
                              "Average Purchase Price": return_average_purchase_df,
                            "Total Purchase Value": return_total_purchase_df})

top_spenders_df.head()

top_spenders_sort_df = top_spenders_df.sort_values("Total Purchase Value", ascending=False)

top_spenders_sort_df["Average Purchase Price"] = top_spenders_sort_df["Average Purchase Price"].map("${:.2f}".format)
top_spenders_sort_df["Total Purchase Value"] = top_spenders_sort_df["Total Purchase Value"].map("${:.2f}".format)

top_spenders_sort_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items_df = pymoli_purchases_df.groupby(["Item ID", "Item Name"])

popular_items_df.head()

pymoli_items_df = popular_items_df["Price"].mean()

pymoli_items_df

popular_total_purchase_df = popular_items_df["Price"].sum()

popular_total_purchase_df

most_popular_items_df = pd.DataFrame({"Purchase Count": popular_items_df['Price'].count(),
                                      "Item Price": pymoli_items_df,
                                      "Total Purchase Value": popular_total_purchase_df})

most_popular_items_df

most_popular_items_df = most_popular_items_df.sort_values(by='Purchase Count', ascending=False)

most_popular_items_df["Item Price"] = most_popular_items_df["Item Price"].map("${:.2f}".format)
most_popular_items_df["Total Purchase Value"] = most_popular_items_df["Total Purchase Value"].map("${:.2f}".format)

most_popular_items_df
```
